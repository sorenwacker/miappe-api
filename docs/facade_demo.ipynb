# Interactive Facade Demo

This notebook demonstrates the interactive facade for creating MIAPPE and ISA entities with tab completion and help functionality.

## 1. Getting Started

The facade provides a user-friendly interface that works well in Jupyter notebooks and IPython with:
- Tab completion for entity names and methods
- Built-in help for field information
- Example code generation
- Field search functionality

```python
from metaseed import miappe, isa, ProfileFacade
```

## 2. Creating a Profile Facade

Use the convenience functions `miappe()` and `isa()` or create a `ProfileFacade` directly.

```python
# Convenience functions
m = miappe()
print(m)

# Or with explicit version
m = miappe(version="1.1")
print(m)

# Or using ProfileFacade directly
m = ProfileFacade("miappe", "1.1")
print(m)
```

## 3. Discovering Entities

List all available entities in the profile.

```python
# List all entities
print("Available entities:")
for entity in m.entities:
    print(f"  - {entity}")
```

```python
# Profile overview with help()
m.help()
```

## 4. Entity Helpers

Access entity helpers via attribute access. In Jupyter/IPython, use Tab to see available entities.

```python
# Access an entity helper (use Tab completion: m.<Tab>)
inv = m.Investigation
print(inv)

# Case-insensitive access also works
inv2 = m.investigation
print(inv2)
```

## 5. Field Information

Get detailed information about entity fields.

```python
# Show all field information
m.Investigation.help()
```

```python
# Get required fields
print("Required fields:")
print(m.Investigation.required_fields)

print("\nOptional fields:")
print(m.Investigation.optional_fields)
```

```python
# Get detailed info for a specific field
print("unique_id field info:")
print(m.Investigation.field_info("unique_id"))

print("\ntitle field info:")
print(m.Investigation.field_info("title"))
```

```python
# See which fields contain nested entities
print("Nested entity fields:")
print(m.Investigation.nested_fields)
```

## 6. Example Code Generation

Generate example code for creating entities.

```python
# Generate example code
m.Investigation.example()
```

```python
m.Study.example()
```

## 7. Creating Entities

Create entity instances using the helper.

```python
import datetime

# Create an investigation using .create()
investigation = m.Investigation.create(
    unique_id="INV-001",
    title="Drought Stress Study in Wheat",
    description="Investigating drought response in wheat varieties",
    submission_date=datetime.date(2024, 1, 15),
)

print(f"Created: {investigation.title}")
print(f"ID: {investigation.unique_id}")
```

```python
# Or call the helper directly (shorthand)
study = m.Study(
    unique_id="STU-001",
    title="Field Trial 2024",
    description="Multi-location field trial",
)

print(f"Created: {study.title}")
```

## 8. Search Functionality

Search for entities or fields by keyword.

```python
# Search for entities or fields containing "date"
results = m.search("date")
print("Fields containing 'date':")
for r in results:
    print(f"  {r}")
```

```python
# Search for "sample"
results = m.search("sample")
print("Results for 'sample':")
for r in results:
    print(f"  {r}")
```

## 9. Working with ISA Profile

The same facade works with ISA profile entities.

```python
# Create ISA facade
i = isa()
print(i)

print("\nISA entities:")
for entity in i.entities:
    print(f"  - {entity}")
```

```python
# Get help on ISA Investigation
i.Investigation.help()
```

```python
# Create ISA entities
isa_inv = i.Investigation(
    identifier="MTBLS-001",
    title="Metabolomics Study",
    description="LC-MS metabolite profiling",
)

print(f"ISA Investigation: {isa_inv.title}")
```

## 10. Building Nested Structures

Create complex nested entity structures using the facade.

```python
# Create a complete investigation with nested studies
full_investigation = m.Investigation(
    unique_id="INV-COMPLETE",
    title="Complete Phenotyping Project",
    description="A comprehensive phenotyping investigation",
    studies=[
        m.Study(
            unique_id="STU-001",
            title="Greenhouse Trial",
            description="Controlled environment study",
        ),
        m.Study(
            unique_id="STU-002",
            title="Field Trial",
            description="Multi-location field experiment",
        ),
    ],
)

print(f"Investigation: {full_investigation.title}")
print(f"Number of studies: {len(full_investigation.studies)}")
for study in full_investigation.studies:
    print(f"  - {study.title}")
```

## Summary

The interactive facade provides:

1. **Tab completion**: Access entities with `profile.<Tab>`
2. **Help system**: `entity.help()` for field documentation
3. **Field discovery**: `required_fields`, `optional_fields`, `nested_fields`
4. **Example generation**: `entity.example()` for code templates
5. **Search**: `profile.search("keyword")` to find entities/fields
6. **Creation**: `entity.create()` or `entity()` to instantiate

This makes it easy to explore and use the MIAPPE and ISA profiles without memorizing the schema.
